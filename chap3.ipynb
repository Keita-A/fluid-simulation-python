```python
import numpy as np
import matplotlib.pyplot as plt
```

3.2

```python
c = 1
dt = 0.5
dx = 1

n_max = 20
t_max = 10

x = np.linspace(0, (n_max-1)*dx, n_max)
q = np.zeros(n_max)

q[0:int(n_max/2)] = 1.0

plt.plot(x, q, marker='o', lw=2, label=f't = {0}')

for t in range(1, t_max+1):
    qbar = q.copy()
    for j in range(1, n_max-1):
        qbar[j] = q[j] - dt * c * (q[j+1] - q[j]) / dx 
    
    for j in range(1, n_max-1):
        q[j] = 1/2 * ((q[j] + qbar[j]) - dt * c * (qbar[j] - qbar[j-1])/dx)

    
    if (t % 2 == 0):
        plt.plot(x, q, marker='o', lw=2, label=f't = {t}')

plt.grid()
plt.legend(loc='upper right')
plt.show()
```

```python
c = 1
dt = 0.05
dx = 0.1

n_max = 100
t_max = 48

x = np.linspace(0, (n_max-1)*dx, n_max)
q = np.zeros(n_max)

q[0:int(n_max/2)] = 1.0

plt.plot(x, q, marker='o', lw=2, label=f't = {0}')

for t in range(1, t_max+1):
    qold = q.copy()

    R = np.zeros(q.shape)
    for i in range(1, n_max-1):
        f_right = 1/2 * (c * qold[i+1] + c * qold[i] - abs(c) * (qold[i+1] - qold[i]))
        f_left = 1/2 * (c * qold[i] + c * qold[i-1] - abs(c) * (qold[i] - qold[i-1]))
        R[i] = (f_right - f_left) / dx
    
    dq = np.zeros(q.shape)
    c_p = 1/2 * (c + abs(c))
    c_n = 1/2 * (c - abs(c))
    for i in range(1, n_max-1):
        dq[i] = (-dt * R[i] + dt/dx*c_p * dq[i-1]) / (1 + dt/dx*abs(c))
    
    for i in range(n_max-2, 0, -1):
        dq[i] = dq[i] - (dt/dx*c_n) * dq[i+1] / (1 + dt/dx*abs(c))
    
    for i in range(0, n_max):
        q[i] = qold[i] + dq[i]
    
    if t % 16 == 0:
        plt.plot(x, q, marker='o', lw=2, label=f't = {t}')


plt.grid()
plt.legend(loc='upper right')
plt.show()
```
